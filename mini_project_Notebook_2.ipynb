Analysis 2 (Multivariate – Multiple Variables)

What Information You Seek


The purpose of this analysis is to identify and understand the relationships between multiple variables present in the student performance dataset. By analyzing how different features such as study time, failures, absences, parental education, and other factors interact with each other, we aim to determine whether any strong associations exist that influence student performance. This helps in understanding how one factor, such as study time or attendance, may impact academic results like grades (G1, G2, G3). Additionally, it is useful for identifying dependent and independent relationships among features, which is important for selecting relevant attributes and improving machine learning model performance. The goal is to detect patterns, dependencies, and possible correlations that can contribute to better prediction of student academic outcomes.

Type of EDA

This analysis is classified as Multivariate Non-Graphical Analysis because it involves examining multiple variables of the student performance dataset simultaneously without using graphical representations. The analysis is performed using statistical techniques such as correlation matrices (`df.corr()`), which provide numerical values indicating the strength and direction of relationships between variables like study time, failures, absences, and student grades (G1, G2, G3). This method helps in quantitatively measuring how strongly different factors are related to academic performance, allowing us to identify important features that influence student outcomes.

Observation

From the correlation analysis, it is observed that certain features in the student performance dataset show a noticeable relationship with one another. For example, study time and grades (G1, G2, G3) tend to have a positive correlation, indicating that students who study more generally achieve better scores. Similarly, failures show a negative correlation with grades, meaning that students with more past failures tend to have lower academic performance. Absences may also show a slight negative relationship with grades, suggesting that irregular attendance can impact performance. The correlation values indicate the strength of these relationships, where values closer to +1 or -1 represent strong relationships, and values near 0 indicate weak or no relationship. Overall, the dataset shows moderate correlations among some features, suggesting that certain variables do influence student performance while others remain relatively independent.

Conclusion

Based on the multivariate analysis, it can be concluded that there are meaningful relationships between several features in the student performance dataset. These correlations indicate that factors such as study time, failures, and absences have an influence on student grades and overall academic performance. Identifying such relationships is important for building efficient machine learning models, as relevant features can improve prediction accuracy. However, strong correlations between similar variables (such as G1, G2, and G3) should be handled carefully to avoid redundancy. Overall, the dataset provides valuable insights into how different factors interact and affect student performance, making it suitable for further modeling and predictive analysis.

PART 2: GRAPHICAL ANALYSIS (2 REQUIRED)

Analysis 3 (Univariate Graph)

```python
import matplotlib.pyplot as plt

import matplotlib.pyplot as plt

df['G3'].hist()
plt.title("Distribution of Final Grades (G3)")
plt.xlabel("Final Grade")
plt.ylabel("Frequency")
plt.show()
```

What Information You Seek

The main objective of this analysis is to visually understand the distribution of student performance in the dataset. By plotting a histogram, we aim to observe how frequently different grade ranges occur. This helps in identifying the central concentration of marks, the spread of student performance, and whether the results are evenly distributed or skewed. Visualizing the distribution also allows us to detect any unusual patterns such as very high or very low scores (outliers) or gaps in performance levels. Understanding the distribution of student grades is important because it provides insight into overall academic performance and helps in preparing the data for further analysis or machine learning models.

Type of EDA

This analysis is categorized as Univariate Graphical Analysis because it focuses on a single variable (student final grades) and uses a graphical method (histogram) to represent its distribution. Graphical analysis helps in better understanding the data visually compared to numerical summaries. A histogram is particularly useful for showing how student performance is distributed across different grade intervals.

Observation

From the histogram, it is observed that the student grades are mostly concentrated within a specific range, indicating that extremely high or low scores are less frequent. The bars in the histogram show that most students have marks around the average range, with fewer students scoring at the extremes. The distribution appears relatively smooth without major irregularities, suggesting consistent performance among most students.

Conclusion

Based on the graphical analysis, it can be concluded that the student grade data follows an approximately normal (bell-shaped) distribution. Most of the values are clustered around the average, with gradual decreases towards the lower and higher ends. This type of distribution is useful for many machine learning models, as it indicates balanced and well-structured data. Therefore, the student performance feature is suitable for further analysis and predictive modeling.

Analysis 4 (Multivariate Graph)

```python
import matplotlib.pyplot as plt

plt.scatter(df['studytime'], df['G3'])
plt.xlabel("Study Time")
plt.ylabel("Final Grade (G3)")
plt.title("Study Time vs Final Grades")
plt.show()
```

What Information You Seek

The main objective of this analysis is to understand the relationship between two variables: study time and student performance. By plotting a scatter plot, we aim to observe how changes in study time affect final grades. This helps in identifying whether increased study time leads to better academic performance.

Type of EDA

This analysis is categorized as Multivariate Graphical Analysis because it involves two variables and uses a scatter plot to visualize their relationship. Scatter plots are useful for identifying correlations, trends, or patterns between variables.

Observation

From the scatter plot, it can be observed that there is a general trend where students who spend more time studying tend to achieve higher grades. However, the points may not form a perfectly straight line, indicating that other factors also influence performance. Some students with lower study time still achieve decent grades, and vice versa, showing variability in the data.

Conclusion

Based on the analysis, there appears to be a positive relationship between study time and final grades, although it is not perfectly strong. This suggests that while study time plays an important role in academic success, other factors such as attendance, support, or individual ability may also impact performance. Therefore, both variables are useful for further analysis and predictive modeling.

```python
import seaborn as sns
import matplotlib.pyplot as plt

numeric_df = df.select_dtypes(include=['int64', 'float64'])

plt.figure(figsize=(10,8))
sns.heatmap(numeric_df.corr(), annot=True)
plt.title("Correlation Heatmap of Student Dataset")
plt.show()
```

What Information You Seek

The main objective of this analysis is to understand the relationships between multiple variables in the student dataset. By using a heatmap, we aim to identify how strongly different features such as study time, absences, and grades are correlated with each other. This helps in determining which factors have the most influence on student performance.


Type of EDA

This analysis is categorized as Multivariate Graphical Analysis because it examines multiple variables simultaneously. A heatmap is used as a graphical representation to show correlation values between variables using color intensity, making it easier to interpret relationships.

Observation

From the heatmap, it can be observed that some variables show strong positive or negative correlations. For example, final grades (G3) may show a positive correlation with study time, indicating that more study leads to better performance. On the other hand, absences may show a negative correlation with grades, suggesting that higher absenteeism can reduce performance. Many other variables may show weak or no correlation, indicating limited impact.

Conclusion

Based on the heatmap analysis, it can be concluded that certain factors significantly influence student performance, while others have minimal effect. Identifying these relationships is useful for feature selection in machine learning models, as highly correlated variables can improve prediction accuracy. Therefore, the heatmap provides valuable insights for further data analysis and modeling.
